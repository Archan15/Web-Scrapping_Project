```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

url = "https://www.imdb.com/chart/top/"
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

# Movie Title
title1 = []
title = soup.find_all('td', class_="titleColumn")
for i in title:
    j = i.find('a').text
    title1.append(j)

# Released Year
year = []
span = soup.find_all('span', class_="secondaryInfo")
for i in span:
    j = i.text.strip('()')
    year.append(j)

# Movie Rating
rating1 = []  
ratings = soup.find_all('td', class_="ratingColumn imdbRating")
for i in ratings:
    j = i.find('strong').text
    rating1.append(j)  # Append the rating values to rating1


# Print Movie Title, Released Year, and Rating side by side
'''print("Movie Title | Released Year | Rating")
for t, y, r in zip(title1, year, rating1):
    print(f"{t} | {y} | {r} â˜…")'''


# Save data to CSV file
'''filename = "top_movies.csv"
with open(filename, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Movie Title", "Released Year", "Rating"])
    for t, y, r in zip(title1, year, rating1):
        writer.writerow([t, y, r])

print(f"Movie data saved to {filename} successfully.")'''


# Create a DataFrame for the movie data
data = {
    "Movie Title": title1,
    "Released Year": year,
    "Rating": rating1
}
df = pd.DataFrame(data)

# Save data to CSV file
filename = "top_movies.csv"
df.to_csv(filename, index=False)

print(f"Movie data saved to {filename} successfully.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

amazon_url = "https://api.scrapingdog.com/scrape?api_key=647cbaedb94fe27c2f5f8100&url=https://www.amazon.in/s?k=iphones&crid=2MIBF168KO3FC&sprefix=iphone%2Caps%2C404&ref=nb_sb_noss_1"
response = requests.get(amazon_url)
response

soup = BeautifulSoup(response.text, 'html.parser')

# Model Name and Colour
model = []
model1 = soup.find_all('span',class_ = "a-size-medium a-color-base a-text-normal")
for i in model1:
    j = i.text.strip('()')
    model.append(j)
print(model)

# Price
price = []
price1 = soup.find_all('span', class_ = "a-price-whole")
for i in price1:
    j = i.get_text()
    price.append(j)
print(price)


# Rating
rating = []
rating1  = soup.find_all('span',class_ = "a-icon-alt")
for i in rating1:
    j = i.get_text()
    rating.append(j)
print(rating)


# Combine all the data Side by side
'''print("Model Name | Price  | Ratings")

for m,p,r in zip(model,price,rating):
    print(f"{m} | {p} | {r}")'''


# Save data to CSV file
filename = "amazon_web.csv"
with open(filename, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Model Name | Price  | Ratings"])
    for m,p,r in zip(model,price,rating):
        writer.writerow([m,p,r])

print(f"Amazon web data saved to {filename} succesfully")

x = pd.read_csv("amazon_web.csv")
print(x)
```
